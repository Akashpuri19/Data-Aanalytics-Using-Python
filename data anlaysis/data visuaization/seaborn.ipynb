```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df =sns.load_dataset('titanic')
```

Visualization on the basis of two values
- categorical data
- numerical data


Basic graph Types
- Line Graph : show trends over time
- Bar Graph : compare values betwween categories(discrete data)
- Pie Chart : compare values between categories( percentage data)
- Histogram : show distribution of numerical data
- Box Plot : statistical distribution of numerical data
- Heat Map : show matrix of vlues using color
- Scatter Plot : show relationship betwwen two variables 
- Bubble Chart : show relationship between three variables

```python
#countplot : Survival based on gender 
sns.countplot(df, x='survived', hue='sex', palette='RdBu')
plt.title('survived vs not survived by gender')
plt.show()
```

```python
sns.barplot(df, x='sex', y= 'fare', hue='class')
plt.title('average fare by passenger class and gender')
```

```python
sns.pointplot(df, x='class', y ='survived', hue ='sex', markers =['o','x'])
plt.title('survival rate by class and gender')
plt.show()
```
